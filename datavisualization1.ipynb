```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mlt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
data = sns.load_dataset('tips')   # Load seaborn dataset
```

```python
data.info()
```

```python
sns.barplot(x='day' , y='tip' , data=data)    # Seaborn barchart
```

```python
sns.barplot(x='day' , y ='total_bill' , data=data , hue ='sex')    # Barplot with multicolumn
```

```python
sns.barplot(x='day' , y ='total_bill' , data=data , hue ='sex' , palette='winter_r')   
```

```python
sns.barplot(x='day' , y ='total_bill' , data=data , hue ='smoker')
```

```python
sns.barplot(x='total_bill' , y ='day' , data=data)    # Barplot vertical
```

```python
sns.barplot(x='day',y='tip',data=data,palette='spring')
```

```python
sns.barplot(x='day',y='tip',data=data,palette='spring',order=['Sat','Sun','Thur','Fri',])  # Barplot with order form
```

```python
from numpy import mean,median
```

```python
sns.barplot(x='day',y='total_bill',data=data,estimator=mean,palette='spring')
```

```python
sns.barplot(x='day',y='total_bill',data=data,estimator=median,palette='spring')
```
