```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mlt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
data = sns.load_dataset('tips')
```

```python
sns.stripplot(x=data['tip'],color='blue')    # Stripplot 
```

```python
sns.stripplot(x=data['total_bill'],color='red')
```

```python
sns.stripplot(x='day',y='total_bill',data=data)
```

```python
sns.stripplot(x='day',y='total_bill',data=data , jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=data , jitter=0.2)
```

```python
sns.stripplot(x='total_bill',y='day',data=data)
```

```python
sns.stripplot(x='total_bill',y='day',data=data ,jitter=1, linewidth=1.2)
```

```python
sns.stripplot(y='total_bill',x='day',data=data ,hue='sex')
```

```python
sns.stripplot(y='total_bill',x='day',data=data ,hue='sex' , jitter=1,split=True)
```

```python
sns.stripplot(y='total_bill',x='day',data=data ,hue='sex',marker='D',size=10,edgecolor='gray',alpha=0.5,split=True)
```

```python
sns.stripplot(y='total_bill',x='day',data=data ,jitter=True)
sns.violinplot(y='total_bill',x='day',data=data,color='0.8')       # Stripplot with violinplot data visualization
```
