```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
data = sns.load_dataset('iris')
```

```python
x = sns.PairGrid(data)
x = x.map(plt.scatter)        # PairGrid and scatter plot data visualization
```

```python
x = sns.PairGrid(data)
x = x.map_diag(plt.hist)
```

```python
x = sns.PairGrid(data)
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
```

```python
x = sns.PairGrid(data,hue='species')
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
```

```python
x = sns.PairGrid(data,hue='species')
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

```python
x = sns.PairGrid(data,hue='species',palette='coolwarm')
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
x = x.add_legend()           
```

```python
x = sns.PairGrid(data,hue='species',palette='autumn')
x = x.map_diag(plt.hist,histtype='step')
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

```python
x = sns.PairGrid(data,vars=['petal_length','petal_width'])
x = x.map(plt.scatter)
```

```python
x = sns.PairGrid(data,vars=['petal_length','petal_width'])
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
```

```python
x = sns.PairGrid(data,vars=['petal_length','petal_width'],hue='species')
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
```

```python
x = sns.PairGrid(data,vars=['petal_length','petal_width'],hue='species')
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

```python
x = sns.PairGrid(data,vars=['petal_length','petal_width'],hue='species')
x = x.map_diag(plt.hist,edgecolor='black')
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

```python
x = sns.PairGrid(data,x_vars=['petal_length','petal_width'],y_vars=['sepal_length','sepal_width'])
x = x.map(plt.scatter)
```

```python
x = sns.PairGrid(data)
x = x.map_diag(plt.hist)
x = x.map_upper(plt.scatter)
x = x.map_lower(sns.kdeplot)
```

```python
x = sns.PairGrid(data,hue='species')
x = x.map_diag(plt.hist)
x = x.map_upper(plt.scatter)
x = x.map_lower(sns.kdeplot)
x = x.add_legend()
```

```python
x = sns.PairGrid(data,hue='species',hue_kws={'marker':['D','s','+']})

x = x.map(plt.scatter)
x = x.add_legend()
```

```python
x = sns.PairGrid(data,hue='species',hue_kws={'marker':['D','s','+']})

x = x.map(plt.scatter , s =30)
x = x.add_legend()
```

```python
x = sns.PairGrid(data,hue='species',hue_kws={'marker':['D','s','+']})

x = x.map(plt.scatter ,s=30 , edgecolor='black')
x = x.add_legend()
```
