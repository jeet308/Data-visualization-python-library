```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mlt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
no = np.random.randn(200)
```

```python
sns.distplot(no)  #   Distplot
```

```python
sns.distplot(no , color='blue')   # Color 
```

```python
lable = pd.Series(no,name='X')
```

```python
sns.distplot(lable)
```

```python
sns.distplot(lable,hist=False)
```

```python
sns.distplot(lable,hist=False,rug=True)
```
